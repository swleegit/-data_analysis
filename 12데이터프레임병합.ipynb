# 데이터 합치기 (concat)

concat 함수는 강의시간에 배운것 처럼 두 데이터프레임을 합치는 역할을 합니다. 주로 데이터프레임 구성이 같을 때 사용하며, 다를 경우 NaN 이 발생할 수 있으므로, 이에 대한 처리가 필요합니다.

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    },
    index=[0, 1, 2, 3],
) 

df2 = pd.DataFrame(
     {
         "A": ["A4", "A5", "A6", "A7"],
         "B": ["B4", "B5", "B6", "B7"],
         "C": ["C4", "C5", "C6", "C7"],
         "D": ["D4", "D5", "D6", "D7"],
     },
     index=[4, 5, 6, 7],
)


df3 = pd.DataFrame(
    {
        "A": ["A8", "A9", "A10", "A11"],
        "B": ["B8", "B9", "B10", "B11"],
        "C": ["C8", "C9", "C10", "C11"],
        "D": ["D8", "D9", "D10", "D11"],
    },
    index=[8, 9, 10, 11],
)
print(df1)
print(df2)
print(df3)
```

```python
# axis=0 으로 df1, df2를 합치기
# axis = 0 : 행 합치기: 행이 같든 안같든 두개의 데이터 프레임 행수의 합 만큼 행이 생성된다.
# default axis = 0이다.
# default join = 'outer' (여기서는 행합치기니까) 열의 이름이 중복되는 것은 한개로 중복되지 않는 것은 각각 : 즉 합집학 개념으로 열이 생성됨
# join = 'inner' 교집합만 표현, 여기서 사용하면 열 중 교집합만 표현함.
pd.concat([df1,df2],axis = 0)  

```

```python
#axis = 1로 df1,df2합치기
#합친다는 것은 안에 있는 value를 합친다는 의미가 아니다. df1 a행b열에 1있고 df2 a행b열에 2있을때 합치면 3이 되는게 아니라
#합친 데이터프레임에 a행 b열이 두개 생긴다. 각각 표현된다. 
#위에서 행과 열의 이름을 숫자의 나열로 해놔서 헷갈릴 수 있는데, 철수 영희 민수 처럼 이름으로 생각해보자.
#그러면 0이라는 이름의 행, D라는 열의 값은 없으니 NaN으로 뜨는 것 
pd.concat([df1,df2], axis =1)
```

```python
# axis=0 으로 df2, df3를 합쳐보세요
pd.concat([df2,df3],axis = 0 )
```

```python
# axis=0 으로 df1, df2, df3를 합쳐보세요
pd.concat([df1,df2,df3],axis = 0)
```

```python
df4 = pd.DataFrame(
    {
        "B": ["B2", "B3", "B6", "B7"],
        "D": ["D2", "D3", "D6", "D7"],
        "F": ["F2", "F3", "F6", "F7"],
    },
    index=[2, 3, 6, 7],
)

df4
```

```python
# df1 과 df4를 axis=1로 합쳐보세요
pd.concat([df1,df4],axis = 1 )
```

```python
# df1 과 df4를 axis=1로 합쳐보세요. 이때 ignore_index=True로 합쳐보시고 위의 결과와 비교해보세요
pd.concat([df1,df4],axis = 1,ignore_index = True ) #axis = 1, ignore_index = True -> 열기준 인덱스 재배열
#합쳤을때는 df1과 df2를 그냥 합친것이기 때문에 index이름 그대로 합친다. 오름 내림 차순으로 재배열 하지 않는다. 
#만약 합쳤을 때 행 인덱스 이름이 1 2 2 6 5 5 이고
#만약 axis = 0, ignore_index = True 이면 행의 이름이 0,1,2,3,4,5 으로 이름이 바뀐다.
```

# 데이터 병합 (merge)


###### 기준이 되는 열을 key로 표현하며, key에 해당하는 열은 반드시 모든 데이터프레임에 존재해야한다.
###### 아래의 실습부분을 차근히 따라하면서 데이터가 어떻게 합쳐지는지 눈으로 확인하세요
###### 병합된 데이터셋 = pd.merge(left_df,right_df,how = 'inner/outer', on = 특정열 열(key))
###### 설정한 특정열(key)를 기준으로 '값'의 합/교집합식으로 합쳐 '행'을 구성됨 
###### 두 데이터 프레임의 열의이름이 같은 경우 각 데이터프레임의 열의 이름을 구분짓는 지시자 left_df는 _x, right_df는 _y가 붙는다
###### 키가 되는 열 : 두 데이터 프레임에 공통으로 존재하는 열 
###### 지시자가 붙은 열 : 두 데이터 프레임에 공통으로 존재하는 열 , 하지만 key가 아니기 때문에 지시자가 붙음
###### 지시자가 없는 열 : 두 데이터 프레임 중 어느하나에만 있는 열 

```python
left = pd.DataFrame(
    {
        "열1": ["a0", "a1", "a2"],
        "열2": ["b0", "b1", "b2"],
        "열3": ["c0", "c1", "c2"],
    }
)

print(left)

right = pd.DataFrame(
    {
        "열1": ["a1", "a3","a4"],
        "열2": ["b2", "b3","b4"],
        "열4": ["d0", "d1","d4"],
    }
)

print(right)
```

```python
# left와 right를 merge로 합쳐보세요. key는 열1 열을 이용하세요 (outer join)
pd.merge(left,right,how = 'outer', on = '열1')
```

```python
# left와 right를 merge로 합쳐보세요. key는 key 열을 이용하세요 (inner join)
pd.merge(left,right,how = 'inner', on = '열1')
```

```python
# left와 right를 merge로 합쳐보세요. key는 key 열을 이용하세요 (left join)
pd.merge(left,right,how = 'left', on = '열1')
```

# 그룹 통계량
###### 데이터 분석 시에 특정 그룹(남/여, 연령대)로 묶어서 각 그룹별 통계치를 확인하고 싶은 경우가 많음
###### pandas에서 groupby라는 함수를 이용하여 특정 칼럼에 대해서 그룹화를 수행할 수 있으며 그룹별 통계량을 산출할 수 있음 

```python
import seaborn as sns
data = sns.load_dataset('titanic')
data.head()
```

```python
groupby_survived = data.groupby(['survived'])
for key, group in groupby_survived:
    print('key ==', key)
    print('=========')
    print(group.loc[:,:'age'].head())
    
#df.groupby([범주형자료형으로 이루어진 열의 이름])
#두가지 변수가 할당되는데 해당 열의 조합으로 나올 수 있는 값과 그 값에 해당하는 원본의 dataset
```

```python
group_0 = groupby_survived.get_group(0) #특정 키 값에 해당하는 그룹화된 데이터 셋을 가져올 수 있음 
group_0.head()
```

```python
group_survived_sex = data.groupby(['survived','sex'])
for key, group in group_survived_sex:
    print('key ==', key)
    print('=========')
    print(group.loc[:,:'age'].head())
    
#key = (0,female), group: (0,female)에 해당하는 dateset
#key = (1,female), group: (0,female)에 해당하는 dateset ... 
```

```python
group_survived_sex.describe()
```

```python
# 각 그룹에 mean, max, min, std를 구해보세요
```

```python
group_survived_sex.agg(['mean', 'max', 'min', 'std'])
```

```python
group_survived_sex.agg({'pclass': 'mean', 
                        'age' : ['mean', 'max', 'min']})
```

```python
group_survived_sex.filter(lambda x: x.age.mean() > 30)
#filter 함수를 이용하여 특정 조건에 해당하는 그룹화 된 데이터 프레임을 선택할 수 있음. 
```
