```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid') #배경 스타일 지정 

fig = plt.figure(figsize=(15,5)) #전체 크기 지정

#그래프 3개를 그리기 위한 객체 3개 생성 및 위치 결정 
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

#seaborn 라이브러리에서 countplot 사용: 빈도수를 막대그래프로 표현
#x축 어떤 열을 나타낼지, data = titanic, titanic 데이터 프레임을 사용한다.
#범주형 자료형 정의 https://m.blog.naver.com/libido1014/120113775017

sns.countplot(x='pclass', data=titanic, ax=ax1)
#등급의 빈도수를 표현 
sns.countplot(x='pclass', hue='sex', data=titanic, ax=ax2)
#어떤 한명에게 부여된 특징 : 성별, 등급 
#그걸 동시에 표현
sns.countplot(x='pclass', hue='sex', dodge=False, data=titanic, ax=ax3)
#중첩해서 표현 
plt.show()
```

## 범주형 데이터를 시각화 하기
#### 빈도그래프 countplot : 범주형 데이터 개수 막대 그래프 (+another 범주형)
#### ex) 1,2,3 등급 (범주형) 남,여(범주형) 1 등급안에 남,여 2등급안에 남,여... 범주형과 범주형 , 그 두가지를 동시에 표현
#### 히트맵 heatmap: 범주형과 범주형끼리 피봇테이블(교집합 경우의수)를 색으로 표현 
#### stripplot : 범주형안에 들어있는 연속형 변수의 분포를 확인 (남자, 여자의 요금) 분포가 달라도 동일한 폭
#### swarmplot :  분포가 min ~ max 가 있음 분포를 알아보기 쉽다.
#### boxplot : 범주형과 연속형의 boxplot 등급별 나이분포 or 하나의 등급별 성별별 나이분포
#### violinplot : 범주형과 범주형 : 정규분포모양
